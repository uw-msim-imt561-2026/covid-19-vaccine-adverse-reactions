# Remove least common symptoms

```python
# get symptom counts
symptom_counts = df_west_filtered['symptom'].value_counts()
df_symptom_counts = symptom_counts.to_frame()
df_symptom_counts = df_symptom_counts.sort_values(by='count', ascending=True)
df_symptom_counts = df_symptom_counts.reset_index()
print(df_symptom_counts.shape)
```

```python
df_symptom_counts.head()
```

```python
# do outlier analysis
# IQR Functions for outliers - created these during lab_02_todo
def get_iqr_upper(x):
    return x.quantile(0.75)+(1.5*(x.quantile(0.75)-x.quantile(0.25)))

def get_iqr_lower(x):
    iqr_low = x.quantile(0.25)-(1.5*(x.quantile(0.75)-x.quantile(0.25)))
    if iqr_low < 0:
        return 0
    else:
        return iqr_low
```

```python
# get lower IQR threshold
lower = get_iqr_lower(df_symptom_counts['count'])
upper = get_iqr_upper(df_symptom_counts['count'])
print(lower)
print(upper)
```

```python
# Get symptoms where unique symptom count <= 3
symptom_counts1 = df_west_filtered['symptom'].value_counts()[df_west_filtered['symptom'].value_counts() <= 25]
print("Number of rows with 2 or fewer instances of a symptom: ", len(symptom_counts1))
```

```python
# get least common symptoms to drop
df_symptom_counts1 = symptom_counts1.to_frame()
df_symptom_counts1 = df_symptom_counts1.sort_values(by='count', ascending=True)
df_symptom_counts1 = df_symptom_counts1.reset_index()
df_symptom_counts1.head()
```

```python
# get list of least common symptoms to drop
symptoms_to_drop = df_symptom_counts1['symptom'].to_list()
print(symptoms_to_drop)
```

```python
# get indices of least common symptoms to drop
idx_to_drop2 = df_west_filtered[(df_west_filtered['symptom'].isin(symptoms_to_drop))].index
print(len(idx_to_drop2))
```

```python
# drop rows with least common symptoms
df_west_filtered = df_west_filtered.drop(idx_to_drop2)
print(df_west_filtered.shape)
```
