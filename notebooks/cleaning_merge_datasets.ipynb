# Data Cleaning: Merge Datasets
This notebook is a scaffold for our data.py script. It will pull dataframes from the other "cleaning" Jupyter notebooks and merge them.

```python
import pandas as pd
import numpy as np
import scipy.stats as sci
import datetime
```

# Read in cleaned csv files - WITH PATHS DEFINED IN CLEANING SCRIPTS, USE THIS SECTION WHEN FILLED IN

## Read in cleaned csv files - WITH LOCAL DATA FILES, DO NOT USE IN FINAL SCRIPTS

```python
# make dataframes
df_VAERS_cl = pd.read_csv('data/VAERSDATA_cleaned.csv')
df_symptoms_cl = pd.read_csv('data/VAERSSYMPTOMS_cleaned.csv')
df_vax_cl = pd.read_csv('data/VAERSVAX_cleaned.csv')
```

```python
# view df_VAERS_cl
df_VAERS_cl.head()
```

# Drop Unnamed Columns

```python
# drop unnamed columns
df_VAERS_cl = df_VAERS_cl.drop('Unnamed: 0', axis=1)
df_symptoms_cl = df_symptoms_cl.drop('Unnamed: 0', axis=1)
df_vax_cl = df_vax_cl.drop('Unnamed: 0', axis=1)
```

# Merge all 3 datasets together

```python
print("Shape of df_VAERS_cl: ", df_VAERS_cl.shape)
print("Shape of df_symptoms_cl: ", df_symptoms_cl.shape)
print("Shape of df_vax_cl: ", df_vax_cl.shape)
```

```python
# merge df_VAERS_cl and df_vax_cl
df_merged_VAERS_vax = pd.merge(df_VAERS_cl, df_vax_cl, how='left', on='VAERS_ID')
print("Shape of df_merged_VAERS_vax: ", df_merged_VAERS_vax.shape)
```

```python
# merge df_symptoms_cl with previous merge
df_merged_ALL = pd.merge(df_merged_VAERS_vax, df_symptoms_cl, how='left', on='VAERS_ID')
print("Shape of df_merged_ALL: ", df_merged_ALL.shape)
```

# Filter Datasets for only Pfizer and Moderna rows

```python
# get indices of rows to drop
idx_to_drop = df_merged_ALL[(df_merged_ALL['VAX_MANU'] != 'PFIZER\\BIONTECH') & (df_VAERS_cl['VAX_MANU'] != 'MODERNA')].index

# sanity check
print(idx_to_drop)
print("Number of non-Pfizer and non-Moderna rows: ", len(idx_to_drop))
print("Total current rows in merged dataframe: ", len(df_merged_ALL))
```

```python
# additional sanity check (put in appropriate index numbers from result above)
# print(df_merged_ALL.iloc[861])
# print(df_merged_ALL.iloc[1096])
```

```python
# drop rows
df_all_filtered = df_merged_ALL.drop(idx_to_drop)

# sanity check
print(df_all_filtered.head(10))
print("Shape of df_merged_ALL: ", df_merged_ALL.shape)
print("Shape of df_all_filtered: ", df_all_filtered.shape)
print(df_all_filtered['VAX_MANU'].unique().tolist()) # verify only Pfizer & Moderna rows remaining
```
