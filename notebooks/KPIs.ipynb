```python
def get_total_events_kpi(df: pd.DataFrame) -> str:
    df_kpi_grouped = (df.groupby(by=["MONTH_YEAR"]).agg(report_count=("VAERS_ID", "count"))).reset_index()
    total_reports = df_kpi_grouped['report_count'].sum()
    total_reports_str = str(total_reports)
    return total_reports_str

# could show rate of change from earliest date in the range to latest date in the range
```

## Percent change total

```python
# get preattentives for total_adverse_events

# get total adverse events in first month of the range
first_month = df_kpi_grouped.at[0,'report_count']
print(first_month)

# get total adverse events in last month of the range
last_month = df_kpi_grouped.at[(len(df_kpi_grouped)-1), 'report_count']
print(last_month)

# calculate percent change
def percent_change (val1, val2):
    percent_change = ((val2 - val1) / abs(val1))*100
    return percent_change
```

```python
pct_change = ((last_month - first_month) / abs(first_month))*100
```

```python
print(pct_change)
```

```python
def get_percent_change_total(df: pd.DataFrame) -> float:
    df_kpi_grouped = (df.groupby(by=["MONTH_YEAR"]).agg(report_count=("VAERS_ID", "count"))).reset_index()
    first_month = df_kpi_grouped.at[0,'report_count']
    last_month = df_kpi_grouped.at[(len(df_kpi_grouped)-1), 'report_count']
    if first_month == 0:
        pct_change = '0 events reported in first month of date range.'
    else:
        pct_change = ((last_month - first_month) / abs(first_month))*100
        pct_change = round(pct_change, 2)
    return pct_change
```

```python
get_percent_change_total(df)
```
